```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

for target_name in ["window_size", "epochs", "learning_rate", "batch_size", "dataset_size"]:
    path = 'results/results_{}.csv'.format(target_name)
    data = pd.read_csv(path, index_col= False)
    x_axis = data[target_name]
    directional_accuracy_of_small_moves = data.directional_accuracy_of_small_moves
    total_directional_accuracy = data.total_directional_accuracy
    directional_accuracy_of_large_moves = data.directional_accuracy_of_large_moves
    plt.figure(figsize=(20,3))
    plt.plot(x_axis, directional_accuracy_of_small_moves, 'b', label='Small Moves Accuracy')
    plt.plot(x_axis, total_directional_accuracy, 'r', label='Total Accuracy')
    plt.plot(x_axis, directional_accuracy_of_large_moves, 'y', label='Large Moves Accuracy')
    plt.xlabel(target_name)
    plt.title("Accuracy")
    plt.legend()
    plt.show()
```
